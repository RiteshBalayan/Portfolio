```python
import pandas as pd
import numpy as np
```

First of all we will create columns for the data, Also its will be helpful to make seprate list of diffrent set of columns in exploratory purpose.

```python
columns = ["event ID", "process ID", "event weight", "MET","METphi"]
events = ["type","E","pt","eta","phi"]
for i in range(1, 22):
    for k in range(1,6):
        columns.append('event' + str(i)+ ' ' +events[k-1])
```

```python
#making specific columns
col_type=[]
col_E=[]
col_pt=[]
col_eta=[]
col_phi=[]
for i in range(1,22):                    #number 22 is just arbitrary, To be on safe-side
        col_type.append('event'+str(i)+ ' type')
        col_E.append('event'+str(i)+ ' E')
        col_pt.append('event'+str(i)+ ' pt')
        col_eta.append('event'+str(i)+ ' eta')
        col_phi.append('event'+str(i)+ ' phi')
```

```python
# Read the data from a file, Remember specific data does not matter in this notebook
data = pd.read_csv("background_chan2b_7.8.csv", names=columns, sep='[,;]', engine='python', header=None)
```

```python
data.head()
```

Replace None with NaN, Just for cautious purpose

```python
def replace_none_with_nan(df):   #function to act on df
    df.replace({None: np.nan}, inplace=True)
    return df
```

```python
replace_none_with_nan(data) #replacng None with NaN
```

# Fill NaN in event type to n "None"

```python
for k in col_type:
    data.loc[:, [k]]=data.loc[:, [k]].fillna('n')
```

# Replace all Other NaN with zero (float)

```python
data.fillna(0, inplace=True)
```

# Taking Log of energy and momentum

```python
for k in col_E:
    data[k] = np.where(data[k] > 0, np.log1p(data[k]), 0)
```

```python
for k in col_pt:
    data[k] = np.where(data[k] > 0, np.log1p(data[k]), 0)
```

```python
data['MET'] = np.where(data['MET'] > 0, np.log1p(data['MET']), 0)
```

# We will Count each type of events in every row and add a new column

```python
def count_n(row):   #these 6 are function to count particular event type
    return row.astype(str).str.count('n').sum()
```

```python
def count_e(row):
    return row.str.count('e-').sum() + row.str.count('e+').sum()
```

```python
def count_mu(row):
    return row.str.count('m-').sum() + row.str.count('m+').sum()
```

```python
def count_b(row):
    return row.astype(str).str.count('b').sum()
```

```python
def count_j(row):
    return row.astype(str).str.count('j').sum()
```

```python
def count_g(row):
    return row.astype(str).str.count('g').sum()
```

```python
data['n_count'] = data[col_type].apply(count_n, axis=1) #following 6 codes are function applied
```

```python
data['e_count'] = data[col_type].apply(count_e, axis=1)
```

```python
data['mu_count'] = data[col_type].apply(count_mu, axis=1)
```

```python
data['jet_count'] = data[col_type].apply(count_j, axis=1)
```

```python
data['bjet_count'] = data[col_type].apply(count_b, axis=1)
```

```python
data['photon_count'] = data[col_type].apply(count_g, axis=1)
```
