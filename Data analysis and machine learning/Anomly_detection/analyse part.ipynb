From Now on We be intrusted in pinning out particualr anomolous events

Folliwing is to find Binary cross entry of indivisual data sets.

```python
def binary_crossentropy (img_in, img_out):
    assert img_in.shape == img_out.shape
    eps = np.finfo(float).eps
    img_out = np.clip(img_out, eps, 1. - eps)
    #return ((img_in - img_out)**2)
    return      - (img_in * np.log(img_out) + (1 - img_in) * np.log(1 - img_out))

# this function can be converted into Mean square error by activation the first comment command. (But I don't found it helpful)
```

This function will give a single number, Loss - to every single data.( Previously we found the loss of whole data), we are intusted in distribution of loss now.

```python
def difference (img_in, img_out):
    return binary_crossentropy(img_in, img_out).mean(axis=(1))
```
